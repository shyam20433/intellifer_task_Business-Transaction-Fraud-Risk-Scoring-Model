```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dataset.csv")
df.head(10)
```

```python
#identifying the columns incase for snakecasing
df.columns
```

```python
#snake casing 
#converting all the features into lowercase ,removing leading blank space and replaced blank space between words by "_"
df.columns=df.columns.str.lower().str.strip().str.replace(" ","_")
df.columns
```

```python
#description of the data

print("shape of the data ",df.shape )
print()
print()
print("data types of each feature ",df.dtypes)
print()
print()
print("missing values of each feature in the dataset",df.isnull().sum())
print()
print()
print("duplicate rows:", df.duplicated().sum())
print()
print()
print("class distribution:\n", df['isfraud'].value_counts(normalize=True))
```

```python
#statistical measure

df.describe(include="all")
```

```python
#since there no missing values we can proceed with out next step
```

```python
#add balance difference
#original balance show how much the sender have after the transaction
#dest balance shows how much the amount added to recieve after the transaction
df["original_balance_difference"]=df["oldbalanceorg"]-df["newbalanceorig"]
df["dest_balance_diff"]=df["newbalancedest"]-df["oldbalancedest"]
df.head()
```

```python
#drop unwanted columns
df=df.drop(columns=["nameorig","namedest","isflaggedfraud"])
df.head(5)
```

```python
#this shows how big the transaction is compared to senders balance
# if amount if 5000 and senders balance is 5000 means the ration is about ~1 so very suspicious
df["amount_to_origin_ratio"] = df["amount"] / (df["oldbalanceorg"] + 1)
```

```python
#this shows how big the transaction is compared to recivers balance
df["amount_to_dest_ratio"] = df["amount"] / (df["oldbalancedest"] + 1)
```

```python
df.tail()
```

```python
for i in df.columns:
    print(i)
```

```python
#converting the step into hours 
df["transaction_hour"]=df["step"]%24
df["transaction_hour"].unique()
```

```python
#most of the fraud transactions done during 12 am to 6 am
df["night_transaction"]=df["transaction_hour"].apply(lambda x : 1 if x<6 else 0)
df=df.drop(columns=["step"])
```

```python
df["night_transaction"].unique()
```

```python
df.head()
```

```python
df["type"].unique()
```

```python
#one hot encoding to avoid priority 
df = pd.get_dummies(df, columns=["type"], drop_first=True)
df.columns=df.columns.str.lower()
df.head(5)
```

```python
#converting the dataset into cleaned one for modeling 
df.to_csv("cleaned_data.csv",index=False)
```
