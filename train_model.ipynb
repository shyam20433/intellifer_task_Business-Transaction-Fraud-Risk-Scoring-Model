```python
import pandas as pd
import numpy as np

df=pd.read_csv("cleaned_data.csv")
x=df.drop(columns=["isfraud"])
y=df["isfraud"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45,stratify=y)
```

```python
#----------------logistic regression model----------------#
```

```python
from sklearn.linear_model import LogisticRegression
logistics_regression=LogisticRegression(class_weight="balanced",max_iter=1000)
logistics_regression.fit(x_train,y_train)
```

```python
#logistic regression report
predict_logistic=logistics_regression.predict(x_test)
print(predict_logistic[0])
predict_probability=logistics_regression.predict_proba(x_test)[:, 1]
print(predict_probability[0])
```

```python
print(confusion_matrix(y_test,predict_logistic))
print(classification_report(y_test,predict_logistic))
```

```python
import numpy as np
new_transaction = np.array([[ 
5000, 20000, 45000, 10000, 15000,
    5000, 5000, 0.25, 0.5,
    14, 0, 0, 0, 1, 0    
]])


y=logistics_regression.predict(new_transaction)
print(y[0])
y_risk=logistics_regression.predict_proba(new_transaction)[:,1]
print(y_risk[0])

print("risk level")
def risk_level(score):
    if score < 0.3:
        return "Low Risk"
    elif score < 0.7:
        return "Medium Risk"
    else:
        return "High Risk"
print(risk_level(y_risk[0]))
```

```python
#----------------decision tree classification----------------#
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier(class_weight="balanced",random_state=42)
decision_tree.fit(x_train,y_train)

```

```python
y_predict=decision_tree.predict(x_test)

print(y_predict[0])
```

```python
new_transaction = pd.DataFrame([[ 
    5000, 20000, 45000, 10000, 15000,
    5000, 5000, 0.25, 0.5,
    14, 0, 0, 0, 1, 0
]], columns=x.columns)

predict=decision_tree.predict(new_transaction)
print(predict[0])
predict_prob=decision_tree.predict_proba(new_transaction)[:,1]
print(predict_prob[0])
print("risk level")
def risk_level(score):
    if score < 0.3:
        return "Low Risk"
    elif score < 0.7:
        return "Medium Risk"
    else:
        return "High Risk"
print(risk_level(predict_prob[0]))
```

```python
#metrix 

print(confusion_matrix(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
